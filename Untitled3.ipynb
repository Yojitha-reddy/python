## maximum length

```python
name = "yojitha reddy nareddy".split()
len_name = []
for ele in name:
    len_name.append(len(ele))
max_name = max(len_name)
for i in range(len(name)):# 0 1 2
    if max_name == len(name[i]):
        print(name[i])
```

```python
n = 'Apollo Brix Chichi Krish'.split()
for name in n:
    o_len = len(name)
    set_len = len(set(name)) #apol 4
    if o_len != set_len:
        print(name,end = " ")
```
